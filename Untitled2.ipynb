### Feiertage

```
dfFeiertage = pd.read_csv('Verkehrsaufkommen/feiertage.csv', index_col='Monat')
dfFeiertage = dfFeiertage.T
print(dfFeiertage)
```

```
dfFeiertage.plot(legend=False, figsize=(15,7))
```

```
concretDfFeiertage=dfFeiertage['NECKARSULM 2']
concretDfFeiertage.plot.bar(figsize=(15,7))
print(concretDfFeiertage)
```

### Vergleich

```
result = pd.concat([concretDfWerktage, concretDfFeiertage], axis=1)
print(result)
result.plot.bar(figsize=(15,7))
```
